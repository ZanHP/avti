# 2. Parametri po registrskih obmojih

## 2.2 Poraba goriva in izpust CO2

Zanimo z analizo porabe goriva in izpustom CO2. Predvidevamo, da tam, kjer imajo avtomobili vejo porabo goriva, izpuajo tudi ve CO2.

```python
poraba = tabela[['Registrsko obmocje', 'Poraba goriva', 'CO2', 'ccm']]
```

```python
poraba = tabela[['Registrsko obmocje', 'Poraba goriva', 'CO2', 'ccm']]
```

```python
skupna_po_krajih = round(poraba.groupby('Registrsko obmocje')[['CO2', 'Poraba goriva', 'ccm']].mean(), 3)

poraba_po_krajih = pd.DataFrame(skupna_po_krajih).sort_values('Poraba goriva', ascending=False)
PORABA = poraba_po_krajih
PORABA['CO2%'] = round((PORABA['CO2'] - min(PORABA['CO2'])) / (max(PORABA['CO2']) - min(PORABA['CO2'])), 3)
PORABA['gorivo%'] = round((PORABA['Poraba goriva'] - min(PORABA['Poraba goriva'])) / (max(PORABA['Poraba goriva']) - min(PORABA['Poraba goriva'])), 3)
PORABA['ccm%'] = round((PORABA['ccm'] - min(PORABA['ccm'])) / (max(PORABA['ccm']) - min(PORABA['ccm'])), 3)

PORABA['|ccm% - gorivo%|'] = abs(PORABA['ccm%'] - PORABA['gorivo%'])

PORABA.reset_index()
```

Stolpec z odstotki je dodan, da lahko z grafa razberemo (ne)ujemanje med porabo goriva in izpustom CO2. 


```python
PORABA[['gorivo%', 'CO2%']].plot(kind='bar', color=('red', 'black'))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

Poraba goriva se kar dobro ujema z izpustom CO2.

## 2.3 Veja prostornina, veja poraba

Preverimo, v kolikni meri dri intuitivna misel, da veja prostornina pomeni vejo porabo. To bomo ugotovili, e izraunamo razliko med odstkovno prostornino in porabo goriva. Naj bi torej veljalo, da tam, kjer imajo avtomobili vejo prostornino, imajo tudi vejo porabo.

```python
razlika = PORABA['|ccm% - gorivo%|']
razlika.sort_values(ascending=False).plot(kind='bar')
```

```python
izracun = pd.DataFrame(razlika).sort_values('|ccm% - gorivo%|', ascending=False).reset_index()

# izpustimo najveje in najmanje odstopanje
za_izracun = izracun[1:-1]

round(za_izracun['|ccm% - gorivo%|'].mean(), 3)
```

Povpreno odstopanje nam da idejo o veljavnosti. Vidimo, da to zelo pogosto ne pomeni prav veliko.

## 2.4 V glavnem mestu se je treba napihovati

Po svetu velja, da so ljudje iz vejih mest bolj naduti. Ali to pomeni, da imajo tudi moneje avtomobile?

```python
moc = tabela[['Registrsko obmocje', 'Moc', 'ccm']]
moc_kraji = round(moc.groupby('Registrsko obmocje')[['Moc', 'ccm']].mean(), 3)
MOC = pd.DataFrame(moc_kraji).sort_values('Moc', ascending=False)

MOC['Moc%'] = round((MOC['Moc'] - min(MOC['Moc'])) / (max(MOC['Moc']) - min(MOC['Moc'])), 3)
MOC['ccm%'] = round((MOC['ccm'] - min(MOC['ccm'])) / (max(MOC['ccm']) - min(MOC['ccm'])), 3)

MOC['|Moc% - ccm%|'] = abs(MOC['Moc%'] - MOC['ccm%'])

MOC.reset_index()
```

```python
MOC[['Moc%', 'ccm%']].plot(kind='bar', color=('yellow', 'brown'))
```

Spoznamo, da v Sloveniji ta hipoteza skoraj velja. 

Zanima nas e, kako sta med seboj odvisna mo in prostornina. 

```python
round(MOC['|Moc% - ccm%|'].mean(), 3)
```

Vidimo, da sta ta dva parametra kar dobro usklajena.

# 3. e nekaj zanimivosti

## 3.1 Za ve moi je potrebnega ve goriva

Preverili bomo e eno izmed zanimivih hipotez in sicer, ali je za ve moi res treba bolj pogosto obiskovati bencinske rpalke?

```python
po_moci = round(tabela.groupby('Moc')['Poraba goriva'].mean(), 3)
MOC_PORABA = pd.DataFrame(po_moci)
xy = MOC_PORABA.plot(kind='line', color='black')
xy.set_xlabel('KONJSKE MOI')
xy.set_ylabel('LITRI')
```

Torej dri, da imajo v povpreju moneji avtomobili vejo porabo goriva. 
Opazimo pa e velika nihanja v obmoju med 150 in 180 konjskimi momi. Sklepamo, da je to zaradi zelo iroke ponudbe avtomobilov s tako mojo in tudi zaradi izboljevanja izkorienosti goriva skozi leta. S tem mislimo, da ima avtomobil s 150KM iz leta 2000 dosti vejo porabo kot enako moan iz leta 2016.

## 3.2 Kako se z leti izboljuje uinkovitost

Preverimo torej, kako se z vianjem oznake EURO zmanjuje izpust CO2.

Ve o oznaki lahko preberete tu: https://en.wikipedia.org/wiki/European_emission_standards.

```python
tabela = tabela[tabela['Leto izdelave'] > 1991] #omejimo se na obdobje, ko so uvedli EURO 0
po_euro = round(tabela.groupby('Okoljevarstvena oznaka')[['Leto izdelave', 'CO2', 'Poraba goriva']].mean(), 3)
evro = pd.DataFrame(po_euro)

EURO = evro.reset_index()

EURO.index = ["EURO " + str(i) for i in range(0,7)]
EURO

xy = EURO['CO2'].plot(kind='bar')
xy.set_ylabel('CO2 g/km')
```

Opazimo, da se z oznako zniujejo emisije, kot smo tudi priakovali. 

Pri EURO 0 vrednosti odstopajo zaradi nepopolnih podatkov za tako stare avtomobile.

```python
co2_leto = round(tabela.groupby('Leto izdelave')[['CO2', 'Okoljevarstvena oznaka']].mean(), 3)
CL = pd.DataFrame(co2_leto)
CO2 = CL['CO2']
clplt = CO2.plot(kind='line')
clplt.set_ylim([100,300])
clplt.set_ylabel('CO2 g/km')
```

```python
OKOL = CL['Okoljevarstvena oznaka']
OKOL
clplt = OKOL.plot(kind='line', color='green')
clplt.set_ylabel('Oznaka EURO')
```

Tudi zmanjevanje izpustov skozi as je priakovano, oznake pa se seveda primerno zviujejo.
